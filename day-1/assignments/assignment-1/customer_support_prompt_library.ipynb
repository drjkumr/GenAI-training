```python
import anthropic
import json
import textwrap
import pandas as pd
from IPython.display import display, Markdown

client = anthropic.Anthropic()  # Reads ANTHROPIC_API_KEY from environment
MODEL = "claude-sonnet-4-20250514"

def run_prompt(system_prompt: str, user_message: str, max_tokens: int = 512) -> str:
    """Send a prompt to the Claude API and return the text response."""
    message = client.messages.create(
        model=MODEL,
        max_tokens=max_tokens,
        system=system_prompt,
        messages=[{"role": "user", "content": user_message}]
    )
    return message.content[0].text

def show(title: str, response: str):
    display(Markdown(f"### 🤖 {title}\n\n{response}\n\n---"))

print("✅ Setup complete.")
```

## 📊 Evaluation Framework
Each response is scored on three dimensions (1–5 scale):

| Dimension | What it measures |
|---|---|
| **Clarity** | Is the response easy to read and understand? |
| **Accuracy** | Does it correctly address the customer's situation? |
| **Conciseness** | Is it appropriately brief without losing important info? |

```python
EVALUATOR_SYSTEM = """
You are an expert evaluator of customer support responses.
Score the following response on three dimensions, each from 1 (poor) to 5 (excellent):
- Clarity: Is the response clear and easy to understand?
- Accuracy: Does it correctly and completely address the customer's situation?
- Conciseness: Is it appropriately brief while covering the key points?

Return ONLY a valid JSON object with this exact structure:
{"clarity": <int>, "accuracy": <int>, "conciseness": <int>, "notes": "<one sentence observation>"}
No markdown, no explanation — only the raw JSON.
"""

def evaluate_response(template_name: str, user_context: str, bot_response: str) -> dict:
    eval_prompt = f"Customer context:\n{user_context}\n\nAgent response:\n{bot_response}"
    raw = run_prompt(EVALUATOR_SYSTEM, eval_prompt, max_tokens=200)
    try:
        scores = json.loads(raw)
    except Exception:
        scores = {"clarity": 0, "accuracy": 0, "conciseness": 0, "notes": "Parse error"}
    scores["template"] = template_name
    scores["avg"] = round((scores["clarity"] + scores["accuracy"] + scores["conciseness"]) / 3, 2)
    return scores

print("✅ Evaluator ready.")
```

```python
# Run evaluations for all 12 templates
all_responses = [
    ("T1: Order Status",          TEMPLATE_1_USER,  response_1),
    ("T2: Delayed Order",         TEMPLATE_2_USER,  response_2),
    ("T3: Missing Delivery",      TEMPLATE_3_USER,  response_3),
    ("T4: Customs Hold",          TEMPLATE_4_USER,  response_4),
    ("T5: App Login Issue",       TEMPLATE_5_USER,  response_5),
    ("T6: Payment Failure",       TEMPLATE_6_USER,  response_6),
    ("T7: Damaged Product",       TEMPLATE_7_USER,  response_7),
    ("T8: Wrong Item",            TEMPLATE_8_USER,  response_8),
    ("T9: Refund Eligible",       TEMPLATE_9_USER,  response_9),
    ("T10: Outside Policy",       TEMPLATE_10_USER, response_10),
    ("T11: Promo Refund",         TEMPLATE_11_USER, response_11),
    ("T12: Escalated Refund",     TEMPLATE_12_USER, response_12),
]

results = []
for name, ctx, resp in all_responses:
    print(f"Evaluating {name}...")
    scores = evaluate_response(name, ctx, resp)
    results.append(scores)

print("\n✅ All evaluations complete.")
```

```python
# Display evaluation results as a table
df = pd.DataFrame(results)[["template", "clarity", "accuracy", "conciseness", "avg", "notes"]]
df.columns = ["Template", "Clarity (1-5)", "Accuracy (1-5)", "Conciseness (1-5)", "Avg Score", "Evaluator Notes"]

display(Markdown("## 📋 Evaluation Results"))
display(df.style.background_gradient(subset=["Avg Score"], cmap="RdYlGn").set_properties(**{'text-align': 'left'}))
```

```python
# Summary statistics
display(Markdown("## 📈 Summary Statistics"))
summary = df[["Clarity (1-5)", "Accuracy (1-5)", "Conciseness (1-5)", "Avg Score"]].describe().round(2)
display(summary)

best = df.loc[df["Avg Score"].idxmax(), "Template"]
worst = df.loc[df["Avg Score"].idxmin(), "Template"]
overall_avg = df["Avg Score"].mean().round(2)

print(f"\n🏆 Best performing template: {best}")
print(f"⚠️  Needs improvement: {worst}")
print(f"📊 Overall library average score: {overall_avg} / 5.0")
```

## 📝 Evaluation Report

### Methodology
- **Model used**: Claude Sonnet 4 (`claude-sonnet-4-20250514`)
- **Evaluator**: Separate Claude API call with a structured scoring rubric
- **Scoring scale**: 1 (poor) → 5 (excellent) per dimension

### Observations

**Clarity**  
All templates scored high on clarity due to explicit structure instructions in the system prompts (numbered steps, bullet points, word limits). Templates with step-by-step instructions (T5, T8) were especially clear.

**Accuracy**  
Templates with richer contextual input (order IDs, dates, amounts) produced more accurate, contextually relevant responses. Templates without concrete data risk being generic.

**Conciseness**  
Word-limit constraints in the system prompts were effective. Escalation templates (T12) scored slightly lower as they needed more words to be appropriately thorough.

### Recommendations
1. Always pass structured customer data (name, order ID, dates) to maximize accuracy.
2. Add explicit word-count caps to all templates to enforce conciseness.
3. For escalations, allow a slightly higher token budget (150–180 words).
4. Periodically re-evaluate with real customer data to catch drift.
